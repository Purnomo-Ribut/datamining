```python
#Modul Library
import numpy as np
import pandas as pd
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
# #modul library score tingkat akurasi
from sklearn.metrics import accuracy_score
```

```python
url = '/content/drive/MyDrive/datamining/tugas/data/online_classroom_data.csv.xls'
dataset = pd.read_csv(url)
```

```python
dataset
```

```python
dataset.info()
```

```python
dataset = dataset.drop(columns=["Unnamed: 0"])
dataset = dataset.replace(",",".",regex=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = dataset.iloc[:, :-1].astype(float)
```

```python
y = dataset.Approved.astype(int)
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.2, random_state=0)
```

```python
#NAIVE BAIYES
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

#modeling
model=GaussianNB()
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
accuracy =accuracy_score(y_test,y_pred)
print ("Acuracy Naive Baiyes = ", accuracy)
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
# Create KNN classifier
knn = KNeighborsClassifier(n_neighbors = 3)
# Fit the classifier to the data
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
accuracy =accuracy_score(y_test,y_pred)
print ("Acuracy KNN = ", accuracy)
```

```python
#DECISSION TREE
from sklearn.tree import DecisionTreeClassifier
### Applying Decision Tree Classification model
decision_tree_classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 27)
decision_tree_classifier.fit(X_train, y_train)
### Predicting the Test set results
y_pred = decision_tree_classifier.predict(X_test)
accuracy =accuracy_score(y_test,y_pred)
print ("Acuracy Decision Tree = ", accuracy)
```
